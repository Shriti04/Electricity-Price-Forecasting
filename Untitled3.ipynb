```python
import pandas as pd
```

```python
data['time'] = pd.to_datetime(data['time'], utc=True)
data2['dt_iso'] = pd.to_datetime(data2['dt_iso'], utc=True)
```

# Merging the datasets on the common timestamp

```python
file1 = pd.merge(data, data2, left_on='time', right_on='dt_iso', how='inner')
```

```python
file1
```

# Dropping the duplicate 'dt_iso' column after merging

```python
file1.drop(columns=['dt_iso'], inplace=True)
```

# check null values

```python
file1.isnull().sum()
```

```python
file1.drop(columns=["generation hydro pumped storage aggregated"],inplace=True)
```

```python
file1.drop(columns=["forecast wind offshore eday ahead"],inplace=True)
```

# Dropping columns that are less relevant or redundant(these values are mostly zero or missing, so we can dropped.)

```python
columns_to_drop = [ 'generation fossil peat', 'generation fossil oil shale',
                   'generation geothermal', 'generation marine', 'generation hydro pumped storage consumption',
                   'weather_id', 'weather_main', 'weather_description', 'weather_icon']
cleaned_df = file1.drop(columns=columns_to_drop, inplace=True)
```

```python
file1.isnull().sum()
```

# Solve null values

```python
columns_to_fill = [
    'generation biomass',
    'generation fossil brown coal/lignite',
    'generation fossil coal-derived gas',
    'generation fossil gas',
    'generation fossil hard coal',
    'generation fossil oil',
    'generation hydro run-of-river and poundage',
    'generation hydro water reservoir',
    'generation nuclear',
    'generation other',
    'generation other renewable',
    'generation solar',
    'generation waste',
    'generation wind offshore',
    'generation wind onshore',
    'total load actual']

for column in columns_to_fill:
    mean_value = file1[column].mean()
    file1[column].fillna(mean_value, inplace=True)
```

```python
file1.isnull().sum()
```

```python
file1['hour'] = file1['time'].dt.hour
file1['day_of_week'] = file1['time'].dt.dayofweek
file1['month'] = file1['time'].dt.month
```

```python
file1.drop(columns=['time'], inplace=True)
```

```python
file1.head()
```

```python
file1.columns
```

# Splitting the data into features (X) and target (y)

```python
feature_columns = [
    'total load actual',
    'generation fossil gas',
    'generation fossil hard coal',
    'generation solar',
    'generation wind offshore',
    'temp',
    'humidity',
    'wind_speed',
    'clouds_all',
    'hour',
    'day_of_week'
]

target_variable = 'price actual'
```

```python
x = file1[feature_columns]
y = file1[target_variable]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
reg.score(x_train,y_train)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# line plot for average price per hour

```python
plt.figure(figsize=(10, 6))
file1.groupby('hour')['price actual'].mean().plot(kind='line', marker='o', color='blue')
plt.title('Average Price by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()
```

## Heatmap
A heatmap can help visualize the relationship between two categorical variables, such as average price by hour and day of the week.

```python
pivot_table = file1.pivot_table(values='price actual', index='hour', columns='day_of_week', aggfunc='mean')

plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, cmap='coolwarm', annot=True, fmt=".2f")
plt.title('Average Price Heatmap by Hour and Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Hour of the Day')
plt.show()
```
